```python
from numpy import asarray
from numpy.random import randn
from numpy.random import randint
from matplotlib import pyplot
import numpy as np

from tensorflow.keras.layers import Input, Dense, Reshape, Flatten, Dropout, Layer
from tensorflow.keras.layers import BatchNormalization, Activation, Embedding, multiply, Concatenate
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import UpSampling2D, Conv2D
from tensorflow.keras.layers import Conv2DTranspose
from tensorflow.keras.models import Sequential, Model, load_model

latent_dim = 40
class_num = 3
count = 12 * class_num

# create and save a plot of generated images
def save_plot(examples):
	# plot images
  for i in range(count):
    pyplot.imshow(examples[i])
    pyplot.axis('off')
    pyplot.savefig('./drive/MyDrive/PokemonProject/generateResult/result_%d.png' %(i))
    pyplot.show()

def build_generator():
    noise = Input(shape=(latent_dim,)) 
    label = Input(shape=(1,), dtype='int32')
    label_embedding = Embedding(class_num, 128)(label)
    label_embedding = Dense(3* 3* 1)(label_embedding)
    label_embedding = Reshape((3, 3, 1))(label_embedding)

    noise_conv=Dense(3 * 3 * 2048, activation="relu", input_dim=latent_dim)(noise)
    noise_conv=Reshape((3, 3, 2048))(noise_conv)

    merge = Concatenate()([noise_conv, label_embedding])

    gen = Conv2DTranspose(1024, (3,3), strides=(2,2), padding='same')(merge)
    gen = LeakyReLU(alpha=0.2)(gen)
    gen = BatchNormalization()(gen, training = True)

    gen = Conv2DTranspose(512, (3,3), strides=(2,2), padding='same')(gen)
    gen = LeakyReLU(alpha=0.2)(gen)
    gen = BatchNormalization()(gen, training = True)

    gen = Conv2DTranspose(256, (3,3), strides=(2,2), padding='same')(gen)
    gen = LeakyReLU(alpha=0.2)(gen)
    gen = BatchNormalization()(gen, training = True)

    gen = Conv2DTranspose(128, (3,3), strides=(2,2), padding='same')(gen)
    gen = LeakyReLU(alpha=0.2)(gen)
    gen = BatchNormalization()(gen, training = True)

    gen = Conv2DTranspose(64, (3,3), strides=(2,2), padding='same')(gen)
    gen = LeakyReLU(alpha=0.2)(gen)
    gen = BatchNormalization()(gen, training = True)

    # output
    out_layer = Conv2D(3, (1,1), activation='tanh', padding='same')(gen)

    generator = Model([noise, label], out_layer)    

    return generator

generator = build_generator()
generator.load_weights('./drive/MyDrive/PokemonProject/result/generatorModelWeight_250.h5')
generator.compile()

# generate images
labels = []
for i in range(0, 12):
  for j in range(0, class_num):
    labels.append(j)

# generate images
noise = np.random.normal(0, 1, (count, latent_dim))
labels = np.array(labels)
X = generator.predict_on_batch([noise, labels])

# scale from [-1,1] to [0,1]
X = (X + 1) / 2.0
# plot the result
save_plot(X)
```
