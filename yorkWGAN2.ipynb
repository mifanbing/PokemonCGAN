```python
import tensorflow as tf

from tensorflow.keras.layers import Input, Dense, Reshape, Flatten, Dropout, Layer
from tensorflow.keras.layers import BatchNormalization, Activation, Embedding, multiply, Concatenate
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import UpSampling2D, Conv2D
from tensorflow.keras.layers import Conv2DTranspose

from tensorflow.keras import activations


from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras import backend
from tensorflow.keras.initializers import RandomNormal
from tensorflow.keras.constraints import Constraint
from tensorflow.keras.optimizers import RMSprop, Adam

import numpy as np
# import tensorflow as tf
import math
from matplotlib import pyplot
import os

class ClipConstraint(Constraint):
	# set clip value when initialized
	def __init__(self, clip_value):
		self.clip_value = clip_value
 
	# clip model weights to hypercube
	def __call__(self, weights):
		return backend.clip(weights, -self.clip_value, self.clip_value)
 
	# get the config
	def get_config(self):
		return {'clip_value': self.clip_value}
 
def wasserstein_loss(y_true, y_pred):
    return backend.mean(y_true * y_pred)

class DCGAN():
    def __init__(self):
        # Input shape
        self.img_rows = 96
        self.img_cols = 96
        self.channels = 3
        self.img_shape = (self.img_rows, self.img_cols, self.channels)
        self.latent_dim = 30

        self.dirName=  [
                  "./drive/MyDrive/PokemonProject/pokemonDatasetYorkBackground/fire",
                  "./drive/MyDrive/PokemonProject/pokemonDatasetYorkBackground/grass",
                  "./drive/MyDrive/PokemonProject/pokemonDatasetYorkBackground/water",
                  # "./drive/MyDrive/PokemonProject/pokemonDatasetYorkBackground/bug",
                  # "./drive/MyDrive/PokemonProject/pokemonDatasetYorkBackground/fly",
                  #"./drive/MyDrive/PokemonProject/pokemonDatasetYorkBackground/rock"
                  ]

        self.class_num = len(self.dirName)
        self.img_names = []
        self.labels = []
        count = 0

        for folder in self.dirName:
          for imageName in os.listdir(folder):
            path = os.path.join(folder, imageName)
            self.img_names.append(path)
            self.labels.append(count)
          count += 1


        # lr_schedulerD = tf.keras.optimizers.schedules.PolynomialDecay(
        #   0.0005,
        #   decay_steps = 2000,
        #   end_learning_rate = 0.0001,
        #   power = 3.0)
        lr_schedulerG = tf.keras.optimizers.schedules.PolynomialDecay(
          0.0001,
          decay_steps = 2000,
          end_learning_rate = 0.00008,
          power = 3.0)
        optD = Adam(learning_rate = 0.0001)
        optG = Adam(learning_rate = 0.0001)

        base_generator = self.build_generator()
        base_discriminator = self.build_discriminator()

        self.generator = Model(
            inputs=base_generator.inputs,
            outputs=base_generator.outputs)

        self.discriminator = Model(
            inputs=base_discriminator.inputs,
            outputs=base_discriminator.outputs)
        self.discriminator.compile(loss = wasserstein_loss, optimizer = optD)

        frozen_D = Model(
            inputs=base_discriminator.inputs,
            outputs=base_discriminator.outputs)
        for layer in frozen_D.layers:
          if not isinstance(layer, BatchNormalization):
            layer.trainable = False

        z = Input(shape=(self.latent_dim,))
        label = Input(shape=(1,), dtype='int32')

        img = self.generator([z, label])
        valid = frozen_D([img, label])

        self.combined = Model([z, label], valid)
        self.combined.compile(loss = wasserstein_loss, optimizer = optG)
 
    def build_generator(self):
        noise = Input(shape=(self.latent_dim,)) 
        label = Input(shape=(1,), dtype='int32')
        label_embedding = Embedding(self.class_num, 1024)(label)
        label_embedding = Dense(3* 3* 60)(label_embedding)
        label_embedding = Reshape((3, 3, 60))(label_embedding)

        noise_conv=Dense(3 * 3 * 2048, activation="relu", input_dim=self.latent_dim)(noise)
        noise_conv=Reshape((3, 3, 2048))(noise_conv)

        merge = Concatenate()([noise_conv, label_embedding])

        gen = Conv2DTranspose(1024, (3,3), strides=(2,2), padding='same')(merge)
        gen = LeakyReLU(alpha=0.2)(gen)
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(512, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(256, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(128, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)
        gen = BatchNormalization()(gen, training = True)

        gen = Conv2DTranspose(64, (3,3), strides=(2,2), padding='same')(gen)
        gen = LeakyReLU(alpha=0.2)(gen)
        gen = BatchNormalization()(gen, training = True)

        relu_custom = lambda x: activations.relu(x, max_value=1)

        # output
        #out_layer = Conv2D(3, (1,1), activation=relu_custom, padding='same')(gen)
        out_layer = Conv2D(3, (1,1), activation='tanh', padding='same')(gen)

        generator = Model([noise, label], out_layer)    

        return generator

    def build_discriminator(self):

        #in_image = Input(shape=(self.img_rows, self.img_cols, 4))
        img = Input(shape=self.img_shape) 
        label = Input(shape=(1,), dtype='int32')
        label_embedding = Embedding(self.class_num, 256)(label)
        label_embedding = Dense(self.img_cols*self.img_rows)(label_embedding)
        label_embedding = Reshape((self.img_cols,self.img_rows,1))(label_embedding)

        merge = Concatenate()([img, label_embedding])
 
        fe = Conv2D(64, (4, 4), strides=(2,2), padding='same')(merge)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)

        fe = Conv2D(128, (4, 4), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)

        fe = Conv2D(256, (4, 4), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)

        fe = Conv2D(512, (3, 3), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)

        fe = Conv2D(1024, (3, 3), strides=(2,2), padding='same')(fe)
        fe = LeakyReLU(alpha=0.2)(fe)
        fe = BatchNormalization()(fe, training = True)

        # flatten feature maps
        fe = Flatten()(fe)
        # dropout
        fe = Dropout(0.2)(fe)
        # output
        out_layer = Dense(1)(fe)

        model = Model([img, label], out_layer)

        return model

    def load_batch_imgs(self,batch_size,shuffle_ind,round):
        img = []
        y = []
        for i in range(batch_size):
            index = shuffle_ind[round*batch_size+i]
            images = image.load_img(self.img_names[index], target_size=(self.img_rows,self.img_cols))
            x = image.img_to_array(images)
            x = np.expand_dims(x, axis=0)
            img.append(x)
            y.append(self.labels[index])
        x = np.concatenate([x for x in img])
        x = x / 127.5 - 1.0
        y = np.array(y)

        return x, y

    def train(self, epochs, batch_size, save_interval, log_interval):
        half_batch = int(batch_size/2)
        valid = -np.ones((half_batch, 1))
        validG = -np.ones((batch_size, 1))
        fake = np.ones((half_batch, 1))

        sample_num = len(self.img_names)

        bat_per_epo=math.floor(sample_num/batch_size)

        errD1History = []
        errD2History = []
        errGHistory = []  

        for epoch in range(epochs):
            
            for j in range(bat_per_epo):
                for _ in range(1):
                  shuffle_ind=np.random.permutation(sample_num)
                  noise = np.random.normal(0, 1, (half_batch, self.latent_dim))
                  imgs, labels = self.load_batch_imgs(half_batch,shuffle_ind,j)
                  gen_imgs = self.generator.predict([noise, labels])

                  d_loss_real = self.discriminator.train_on_batch([imgs, labels], valid)
                  d_loss_fake = self.discriminator.train_on_batch([gen_imgs, labels], fake)

                for _ in range(2):
                  shuffle_ind=np.random.permutation(sample_num)
                  _, labels = self.load_batch_imgs(batch_size,shuffle_ind,j)
                  noise = np.random.normal(0, 1, (batch_size, self.latent_dim))
                  g_loss = self.combined.train_on_batch([noise,labels], validG)

                print("%d %d [D1 loss: %f, D2 loss: %f] [G loss: %f]" %
                          (epoch, j, d_loss_real, d_loss_fake, g_loss))
                
                if epoch % log_interval == 0 and j == 0:
                  errD1History.append(d_loss_real)
                  errD2History.append(d_loss_fake)
                  errGHistory.append(g_loss)
                  pyplot.plot(errD1History, label = "D1") 
                  pyplot.plot(errD2History, label = "D2")       
                  pyplot.plot(errGHistory, label = "G")
                  pyplot.ylim([-3000, 3000])
                  pyplot.show()

            if epoch % save_interval == 0:
                self.save_imgs(epoch)
                self.saveModel(epoch)

    def save_imgs(self, epoch):
        r, c = 5, self.class_num
        noise = np.random.normal(0, 1, (r * c, self.latent_dim))
        labels =[]
        for i in range(0, r):
          for j in range(0, c):
            labels.append(j)

        labels = np.array(labels)

        gen_imgs = self.generator.predict([noise, labels])

        # Rescale images 0 - 1
        gen_imgs = 0.5 * gen_imgs + 0.5

        fig, axs = pyplot.subplots(r, c)
        cnt = 0
        for i in range(r):
            for j in range(c):
                axs[i, j].imshow(gen_imgs[cnt])
                axs[i, j].axis('off')
                cnt += 1
        fig.savefig("./drive/MyDrive/PokemonProject/result2/result_%d.png" % epoch)
        pyplot.close()

    def loadModel(self):
        self.discriminator.load_weights('./drive/MyDrive/PokemonProject/result2/discriminatormModelWeight.h5')
        self.generator.load_weights('./drive/MyDrive/PokemonProject/result2/generatorModelWeight_900.h5')
        self.combined.load_weights('./drive/MyDrive/PokemonProject/result2/ganModelWeight.h5')

    def saveModel(self, epoch):
        try:
          self.combined.save_weights("./drive/MyDrive/PokemonProject/result2/ganModelWeight.h5")
          self.discriminator.save_weights("./drive/MyDrive/PokemonProject/result2/discriminatormModelWeight.h5")
          self.generator.save_weights("./drive/MyDrive/PokemonProject/result2/generatorModelWeight_%d.h5" % epoch)
        except OSError:
           pass
dcgan = DCGAN()
#dcgan.loadModel()
dcgan.train(epochs=2000, batch_size=64, save_interval=50, log_interval=5)



```
